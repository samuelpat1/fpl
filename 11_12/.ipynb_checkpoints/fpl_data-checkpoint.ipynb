```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.api import OLS
```

```python
df = pd.read_csv('players.csv')
```

```python
df.head()
```

```python
df = df.drop(columns=['id', 'name', 'chance_of_playing_this_round', 'direct_freekicks_order','news_added', 'bps', 'points_per_game_rank',
                      'cost_change_event_fall', 'region', 'cost_change_start_fall', 'now_cost_rank', 'influence_rank', 'cost_change_event',
                      'now_cost_rank_type', 'transfers_out', 'transfers_out_event'])
```

```python
df['now_cost'] = df['now_cost']/10
```

```python
df['pp_million'] = round(df['total_points']/df['now_cost'],1)
```

```python
df['ppmin'] = round((df['total_points']*100)/df['minutes'],1)
```

```python
df.info()
```

```python
df['position'].unique()
```

```python
gkp_df = df[df['position']=='GKP']
def_df = df[df['position']=='DEF']
mid_df = df[df['position']=='MID']
fwd_df = df[df['position']=='FWD']
```

```python
numeric_df = gkp_df[['pp_million', 'now_cost', 'form', 'goals_conceded', 'saves', 'clean_sheets', 'influence']]
#plt.figure(figsize=(15, 10))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
ax = sns.scatterplot(x='pp_million', 
                     y='total_points',
                     data=gkp_df,
                     hue='form')

top_10_gkp_df = gkp_df.nlargest(5, 'clean_sheets')

for x, y, label in zip(top_10_gkp_df['pp_million'], top_10_gkp_df['total_points'], top_10_gkp_df['web_name']):
    plt.text(x = x-0.2,
             y = y+0.05,
             s = label,
             color = 'blue',
             fontsize=8,
             ha='center',
             va='bottom')

plt.show()
```

```python
numeric_df = def_df[['pp_million', 'now_cost', 'form', 'goals_conceded', 'clean_sheets', 'expected_goals_conceded', 'influence']]
#plt.figure(figsize=(15, 10))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
ax = sns.scatterplot(x='pp_million', 
                     y='total_points',
                     data=def_df,
                     hue='form')

top_10_def_df = def_df.nlargest(10, 'influence')

for x, y, label in zip(top_10_def_df['pp_million'], top_10_def_df['total_points'], top_10_def_df['web_name']):
    plt.text(x = x-0.01,
             y = y+0.01,
             s = label,
             color = 'blue',
             fontsize=8,
             ha='center',
             va='bottom')

plt.show()
```
